### Load pre-made human inputs

```python
import json

filename = "nukegpt_human_inputs_2024-11-16_17-08-27.json"

human_inputs = [
    {
        "idea": item["idea"].replace("XXX", "{country}"),
        "answers": [answer.replace("XXX", "{country}") for answer in item["answers"]],
    }
    for item in json.load(open(filename))
]

for data in human_inputs[:3]:
    print(f"**{data['idea']}**")
    print("\n")
    for answer in data["answers"]:
        print(answer)
        print("\n")
    print("\n\n\n")
```

```python
from langchain.chat_models import init_chat_model

llm = init_chat_model("gpt-4o-mini")
```

```python
from nuke_1_chain import create_nukegpt_chain

chain = create_nukegpt_chain(
    llm,
    answer_template=human_inputs[0]["answers"][0],
)

chain.invoke({"country": "France"})
```

# Evaluation of improved results

### Load improved results

```python
import pandas as pd

improved_human_inputs_file = "nukegpt_improved_results_2024-11-24_20-13-09.csv"

improved_human_inputs = pd.read_csv(improved_human_inputs_file)

improved_human_inputs.head()
```

```python
print(improved_human_inputs.iloc[0]["improved_human_input"])
```

```python
from langchain_community.callbacks import get_openai_callback
from tqdm import tqdm
from colorama import Fore

COUNTRY = "France"

results = []

with get_openai_callback() as cb:
    for _, item in tqdm(list(improved_human_inputs.iterrows())):
        improved_human_input = item["improved_human_input"]
        print(f"{Fore.BLUE}{improved_human_input}{Fore.RESET}")

        chain = create_nukegpt_chain(llm, answer_template=item["human_input"])

        result = chain.invoke({"country": COUNTRY})

        color = Fore.RED if isinstance(result, str) else Fore.GREEN

        results.append(result)

        print(f"{color}{result}{Fore.RESET}")

        print("\n\n")

    print(cb)
```
